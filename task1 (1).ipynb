```python
dept_eth_summary = df.groupby(["Department","Ethnicity"]).agg({
    "Age": ["max","min"],
    "Annual Salary": "median"
})
dept_eth_summary
```

```python
output_path = "Employee_Analysis_Output.xlsx"

with pd.ExcelWriter(output_path, engine="openpyxl") as writer:
    df.to_excel(writer, sheet_name="Cleaned_Data", index=False)
    dept_summary.to_excel(writer, sheet_name="Dept_Averages")
    dept_eth_summary.to_excel(writer, sheet_name="Dept_Ethnicity_Stats")

print("Saved to:", output_path)
```
